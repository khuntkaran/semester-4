### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40]
y=["food","travel","acomodetion","shopping"]
c=['r','g','y','m']
ex=[0.0,0.2,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=ex,shadow=True,autopct="%0.2f%%",textprops={"fontsize":15,"color":"b"},labeldistance=1.5)
plt.title("Expensis",)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=["Python","Java","C++","C"]
y=[85,70,60,82]
c=['r','g','y','m']
plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5)
plt.title("Popularity of languages")
plt.xlabel("Language")
plt.ylabel("Rating")
plt.show()
```

```python
x=["Python","Java","C++","C"]
y=[85,70,60,82]
c=['r','g','y','m']
plt.barh(x,y,color=c,align='edge',edgecolor='b',alpha=0.5)
plt.title("Popularity of languages")
plt.xlabel("Rating")
plt.ylabel("Language")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["Python","Java","C++","C"]
y=[85,70,60,82]
c=['r','g','y','m']
plt.bar(x,y,color=c,edgecolor='b',alpha=0.5)
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.title("Popularity of languages")
plt.xlabel("Language")
plt.ylabel("Rating")
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,7]
c=['r','g','y','m','b']
plt.scatter(x,y,color=c,marker="*",linewidths=[10,5,2,9,6])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
